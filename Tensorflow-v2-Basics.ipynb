## Constants, Variables and conversions

```python
# Set a constant value
tensor_20 = tf.constant([[10, 10], [20, 20]])
print(tensor_20)
```

It is a Tensor type of shaoe 2x2 as expected. You can extract the values as a numpy array to work with numpy functions or other functionalities

```python
# Obtener los valores directamente desde una constante de TensorFlow con numpy, sin necesidad de una session
tensor_numpy = tensor_20.numpy()
print(tensor_numpy), print(tensor_numpy.dtype)
```

You can create a Tensor from a numpy array

```python
# Podemos convertir un array de numpy a un tensor de TensorFlow directamente
numpy_tensor = np.array([[23,  4], [32, 51]])
tensor_from_numpy = tf.constant(numpy_tensor)
print(tensor_from_numpy)
```

### VAriables

```python
# DEfine a variable
tf2_variable = tf.Variable([[1., 2., 3.], [4., 5., 6.]])
print("Tensor: ", tf2_variable)
# Extract the value to a numpy variable
variable_np = tf2_variable.numpy()
print("Array: ",variable_np)
```

Modify the value in a varibale

```python
tf2_variable[0, 2].assign(100)
print(tf2_variable)
# Assign and add values to the variable
print(tf2_variable.assign_add([[1, 1, 1], [1, 1, 1]]))
```

## Basic operations with tensors

```python
tensor = tf.constant([[1, 2], [3, 4]])
print(tensor)
tensor2 = tf.constant([[1., 2., 3.], [4., 5., 6.]])
print(tensor2)
```

```python
# Scalar sum
print(tensor + 2)
# Multiply
print(tensor * 5)
```

Apply numpy functions to tensors

```python
np.square(tensor)
```

```python
# Obtener la raíz cuadrada de todos los números de un tensor de TensorFlow
np.sqrt(tensor)
```

### Tensor operations

Some usual functions

```python
tf.size(tensor)
```

```python
# Dot product
print(np.dot(tensor, tensor_20))
# Element-wise addition
print(tf.add(tensor, tensor))
# REshape a tensor
print(tf.reshape(tensor2, (3,2)))
```

```python
# TRanspose a tensor
print(tf.transpose(tensor))
```

```python
tf.concat([tensor, tensor, tensor], axis=0)
```

## Initialize a tensor with random values
